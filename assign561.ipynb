```python
import pandas as pd
import numpy as np

with open('temp.txt') as f: 
    lines = f.readlines()

class AllFunctions:
    def convert_list_of_int(self, lines):
        b=[]
        c=[]

        # creating a list of all the given 1000 lines - string within list within list 
        a = [i.split('\n')[0].strip("()").split("), (") for i in lines]

        # creating a single list with all the strings
        for idx,i in enumerate(a):
            b = b +a[idx]

        #creating a single list with strings converted to integers
        for k in b:
            j = k.split(", ")
            j[0]=int(j[0])
            j[1]=int(j[1])
            c.append(j)
        return c

    def data_split(self, list1):
        l1=[]
        l2=[]
        for i in list1[:1000]:
            l1.append(i)
        for i in list1[1000:]:
            l2.append(i)
        return l1,l2
    
    def mapper(self, data):
        ## Map and combine are in this function only
        lst = []
        for i in data:
            lst.append((int(i[0]/100), i[1]))   
        cd = {}
        print(type(lst),"\n")
        for item in lst:
            if item[0] in cd:
                cd[item[0]] = max(cd[item[0]], item[1])
            else:
                cd[item[0]] = item[1]
        return cd
    
    def sort_func(self,map1,map2):
        lst={}
        for i in sorted (map1) :
            lst[i]=map1[i]
        for item in sorted(map2):
            if item in lst:
                lst[item] =  [map1[i] for i in map1 if i==item]
                lst[item].append(map2[item])
            else:
                lst[item]=map2[item]
        return lst
    
    def partition(self,dict1):
        dict_1={}
        dict_2={}
        for i in dict1:
            if i in range(2010,2016):
                dict_1[i]=dict1[i]
            else:
                dict_2[i]=dict1[i]
        return dict_1,dict_2
    
    def reducer(self,ip):
        new_dict={}
        for i in ip:
            if len(ip[i])>1:
                new_dict[i]=max(ip[i]) 
            else:
                new_dict[i]=ip[i]
        return new_dict
                

## Creating object of the class
mapred = AllFunctions()

## converting the given string data into list of integers
final_data = mapred.convert_list_of_int(lines)

## Split the data into two parts
data1,data2 = mapred.data_split(final_data)

##
map1 = mapred.mapper(data1)
map2 = mapred.mapper(data2)
sorted_data = mapred.sort_func(map1,map2)
reducer_ip1,reducer_ip2 = mapred.partition(sorted_data)
final1=mapred.reducer(reducer_ip1)
final2=mapred.reducer(reducer_ip2)

##combining the output
final={**final1, **final2}

## converting the dictionary into data frame
df=pd.DataFrame.from_dict(final,orient='index')
df.columns=['Max temperature']
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'Year'})
print(df)

# saving the dataframe
df.to_csv('561_Anisha.csv')
```
