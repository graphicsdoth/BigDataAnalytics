# Task1 - Step1

```python
##TASK1 -- STEP 1
def date_return(str1):
  str2=str1.split(" ")
  return str2[1] + " " + str2[2]

## making one rdd with all the columns
data_amazon_rdd_master = amazon_rdd.map(lambda x: (date_return(x[0]), x[1], str(x[2])))

print(data_amazon_rdd_master.take(10))
```

```python
data_amazon_rdd_master.map(lambda xs: [x[2].astype(str) for x[2] in xs])

print(data_amazon_rdd_master.take(10))
```

```python
## Converting the rdd to spark dataframe
# SparkSession.createDataFrame(data, schema=None, samplingRatio=None, verifySchema=True)
deptColumns = ["tweet_created_at","user_screen_name","user_id_str"]

df_final = spark.createDataFrame(data_amazon_rdd_master,schema = deptColumns)
df_final.show(truncate=False)

print((df_final.dtypes))
# val df_final = spark.createDataFrame(data_amazon_rdd_master).toDF("tweet_created_at","user_screen_name","user_id_str")
```

```python
#Finding the users who are active atleast 5 days
#select user id, count(distinct tweet created at)
# df1= df_final.select("user_id_str","user_screen_name","tweet_created_at")

# Finding the distinct days
df_distinct = df_final.distinct()
#print(df_distinct.show(10))

#Now we have to count the users and their active days
distinct_count = df_distinct.groupby(["user_id_str","user_screen_name"]).count().where(col("count") >= 5).collect()
# x.show(10)
# x.where(col("count") >= 5).show(truncate=False)
# df_final.groupBy("tweet_created_at").count().show(10)

# print(type(distinct_count))
```

```python
# Convert list to RDD
rdd = spark.sparkContext.parallelize(distinct_count)

schema = StructType([
    StructField('user_id_str', StringType(), True),
    StructField('user_screen_name', StringType(), True),
    StructField('Count', IntegerType(), True),
])

# Create data frame
Final_distinct = spark.createDataFrame(rdd,schema)
print(Final_distinct.schema)
Final_distinct.drop(col("Count")).show()

Final_distinct=Final_distinct.select("user_screen_name","user_id_str")
Final_distinct.show(10)

# # # Write to CSV
Final_distinct.toPandas().to_csv(r'/content/drive/My Drive/Colab Notebooks/IDS561/Daily_Active_Users_Anisha.csv', index = False)
```

# Task1 - Step2

```python

text_DF1 = spark.read.option("header", "false") \
    .option("delimiter", ",") \
    .option("inferSchema", "true") \
    .csv("/content/drive/MyDrive/Colab Notebooks/IDS561/15902564.txt")

text_DF1.show(20, truncate=False)

# selected_list = df_distinct.groupby(["user_id_str"]).count().where(col("count") >= 5).collect()
```

```python
## Changing bigint to string in the given text file
print(text_DF1.dtypes)
from pyspark.sql.types import StringType
text_DF = text_DF1.withColumn("_c0",text_DF1["_c0"].cast(StringType()))
print(text_DF.dtypes)
```

```python
active_users = Final_distinct.drop("user_screen_name")
active_users.show(20,truncate=False )
```

```python
## Performing the left join
experiment_user = text_DF.join(active_users, text_DF._c0 == active_users.user_id_str ,how='left') 
# experiment_user.show(20)

experiment_user = experiment_user.withColumnRenamed("_c0","User_ID")
experiment_user = experiment_user.withColumnRenamed("user_id_str","Whether_Active")

experiment_user.show()
```

```python
#count of all rows
count_all = experiment_user.count()
#count of only Null rows
count_null = experiment_user.where(experiment_user.Whether_Active.isNull()).count()
```

```python
## Replace the null values with "No" and others as "yes"

from pyspark.sql.types import StringType
from pyspark.sql.functions import udf

def replace_yes(r):
  if r != "No":
    return "Yes"
  else:
    return r

df_replace_yes = udf(replace_yes, StringType())

# experiment_user.na.fill("no",subset=["user_id_str"]).show()

experiment_user = experiment_user.na.fill("No")

# experiment_user.show()

new_df = experiment_user.withColumn("Whether_Active",df_replace_yes(experiment_user.Whether_Active))

new_df.show()
```

```python
## % of active users

print(count_all)
print(count_null)
count_yes = count_all - count_null
print(count_yes)

print("The % of active users is ",(count_yes/count_all)*100)
```
